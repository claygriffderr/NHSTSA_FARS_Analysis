## Feature Engineering

Age of veh:

```python
def ageCalc(row):
    if row['MOD_YEAR'] == 0:
        return 999
    else:
        age = row['YEAR']-row['MOD_YEAR']
        return age 
```

```python
df['VEH_AGE'] = df.apply(ageCalc, axis=1)
```
